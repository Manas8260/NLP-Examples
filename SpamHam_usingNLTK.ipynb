```python
import warnings
warnings.filterwarnings('ignore')
```

```python
import pandas as pd 
```

```python
msgs = pd.read_csv('Data/SMSSpamCollection.txt',sep = '\t',names = ['label','messages'])
```

```python
msgs.head(2)
```

```python
msgs.shape
```

```python
import re
from nltk.stem import PorterStemmer
ps = PorterStemmer()
from nltk.corpus import stopwords
```

```python
corpus = []
for i in range(msgs.shape[0]):
    m = re.sub('[^a-zA-Z]',' ',msgs.messages[i])
    m = m.lower()
    m = m.split()
    m = [ps.stem(w) for w in m if w not in set(stopwords.words('english'))]
    m = ' '.join(m)
    corpus.append(m)
```

```python
len(corpus)
```

# Bag of words

```python
y = msgs.label.replace({'ham':1,'spam':0})
y = y.values
```

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,ytrain,ytest = train_test_split(corpus,y,test_size=0.2,random_state=42)
```

```python
len(Xtrain)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 2500, ngram_range = (1,2))
```

```python
Xtrain = cv.fit_transform(Xtrain).toarray()
Xtest = cv.transform(Xtest).toarray()
```

```python
Xtrain.shape
```

```python
#cv.vocabulary_
```

```python
msgs.label.value_counts()
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_det = MultinomialNB().fit(Xtrain,ytrain)
```

```python
spam_det
```

```python
ypred = spam_det.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score, classification_report

accuracy_score(ytest,ypred)
```

```python
print(classification_report(ytest,ypred))
```

# TF-IDF

```python
from sklearn.model_selection import train_test_split
Xtrain,Xtest,ytrain,ytest = train_test_split(corpus,y,test_size=0.2,random_state=42)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tv = TfidfVectorizer(max_features = 2500,ngram_range = (1,2))
Xtrain = tv.fit_transform(Xtrain).toarray()
Xtest = tv.transform(Xtest).toarray()
```

```python
Xtrain
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
spam_tfidf = MultinomialNB().fit(Xtrain,ytrain)
```

```python
spam_tfidf
```

```python
ypred = spam_tfidf.predict(Xtest)
```

```python
accuracy_score(ypred,ytest)
```

```python
print(classification_report(ytest,ypred))
```
